```python
from spatial_models import *
from spatial_indicators import *
from spatial_plotting import *
```

# Load Snapshots

```python
snapshots1, snapshots2, snapshots3, _ = load_data('Results/spatial_simulation.dat')

_, _, _, snapshots4, snapshots5, snapshots6, *_ = load_data('Results/scale_dependent_model.dat')
```

# Local Positive Feedback Model

```python
plot_snapshots(snapshots1, [2, 1.6, 1.2, 1.1, 1], "R", vmin=0, vmax=10)
```

```python
fig = spatial_indicator_grid(snapshots1, "R", shape=(4,1), return_fig=True)
fig.set_size_inches(3, 6)
fig.align_ylabels()
fig.tight_layout(h_pad=0.5)
```

# Local Facilitation Model

```python
plot_snapshots(snapshots2, [1, 0.6, 0.5, 0.4, 0.3], "b", vmin=-1, vmax=1)
```

```python
fig = spatial_indicator_grid(LocalFacilitationModel().vegetation_cover(snapshots2), "b", shape=(4,1), return_fig=True)
fig.set_size_inches(3, 6)
fig.align_ylabels()
fig.tight_layout(h_pad=0.5)
```
